```python
gtcent = np.load('/scratch/sbanerj/trans-eqtl/simulation/10414_450_10_800_30_100_100_0.01_0.5_0.0_0.6_0.6_4.0_0.1_20_0.02/sim002/tejaas/gtcent_1-4000.npy')
expr = np.load('/scratch/sbanerj/trans-eqtl/simulation/10414_450_10_800_30_100_100_0.01_0.5_0.0_0.6_0.6_4.0_0.1_20_0.02/sim002/tejaas/expr.npy')
```
